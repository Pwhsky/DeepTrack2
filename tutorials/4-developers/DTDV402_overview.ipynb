# Module Basics

## What is a module?

A module in the DeepTrack2 framework is a `.py` source code file 
containing classes and methods which implement functionalities.

Inheritance is used in the design of modules DeepTrack2, and understanding 
which modules are lower level and which are higher level ones.

## How is a module structured?

The first thing in a module is always the documentation for the class which
goes over things like key features, structure, examples.

The first class in a module is typically an **abstract** class. 

This abstract class provides a standardized implementation of core functionality (e.g., data processing) to ensure consistent output formatting across all derived classes in the module. For instance, whether `class A` or `class B` produces the output, the format remains uniform. This **abstract** class often extends functionalities from its lower level parent classes (e.g. the `Aberration` class extends the `Feature` class).

---


## The backbone of DeepTrack2

Located in the `DeepTrack2/deeptrack/backend/` is the source code which forms the
backbone of the framework, the modules are:

- `core.py`: Provides the core DeepTrack2 classes to manage and process data.
    In particular it enables the construction of computational pipelines, dependency tracking, and functionality for citing.
    
- `mie.py`: Provides functions to perform Mie scattering calculations.
    In particular computing coefficients for both spherical and stratified spherical harmonics, and calculating the spherical harmonics of the Mie field with an iterative method.

- `pint_definition.py`: Extends Pint's default definitions by introducing project-specific constants
    and unit modification for flexible calculations in he context of DeepTrack2.


- `polynomials.py`: Provides a set of functions which compute Bessel and Riccati-Bessel polynomials and     their derivatives.


- `units.py`: Provides unit management and conversions for simulations involving voxel dimensions and   scaling factors.

## Features

#### An **abstract** class in action

```python
# We want the abstract class to process our data
# in the the same way for all cases.
class AbstractClass:
    """Processes a string to upper case and wraps into a list."""

    def __init__(self, data, **kwargs):
        # As this class gets initiated, it will process the data.
        self.data = data
        self.process_data()

    def process_data(self):
        self.data = [self.data.upper()]
        print(self.data)

# Define a class which will add " bar" to our data.
class Adding(AbstractClass):
    """Adds chars to the data."""

    def __init__(self, data, **kwargs):
        added_data = data + " bar"
        super().__init__(added_data, **kwargs)

# Define a class which will reverse the data.
class Reversing(AbstractClass):
    """Reverses the data."""

    def __init__(self, data, **kwargs):
        reversed_data = data[::-1]
        super().__init__(reversed_data, **kwargs)

# Test the functionality.
Adding(data="foo")
Reversing(data="raboof")
```
